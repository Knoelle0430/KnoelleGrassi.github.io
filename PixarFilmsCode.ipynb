```python
#Correlation Analysis
numeric_cols = [
    'run_time', 'budget_clean', 'box_office_us_canada_outlier', 'box_office_other',
    'box_office_worldwide', 'rotten_tomatoes_score',
    'rotten_tomatoes_counts', 'metacritic_score', 'metacritic_counts',
    'imdb_score', 'imdb_counts', 'budget_clean'
]
corr_matrix = pixar[numeric_cols].corr()
corr_matrix
```

```python
#visualization of Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Pixar Film Variables", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#correlation with worldwide box office
corrs = corr_matrix['box_office_worldwide'].sort_values(ascending=False)
corrs.drop('box_office_worldwide').plot(kind='barh', figsize=(10,6), title="Correlation with Worldwide Box Office")
plt.xlabel("Correlation coefficient")
plt.tight_layout()
plt.show()
```

```python
#correlation with run time
corrs = corr_matrix['run_time'].sort_values(ascending=False)
corrs.drop('run_time').plot(kind='barh', figsize=(10,6), title="Correlation with Run Time")
plt.xlabel("Correlation coefficient")
plt.tight_layout()
plt.show()
```

```python
#correlation with rotten tomatoes score
corrs = corr_matrix['rotten_tomatoes_score'].sort_values(ascending=False)
corrs.drop('rotten_tomatoes_score').plot(kind='barh', figsize=(10,6), title="Correlation with Rotten Tomatoes Score")
plt.xlabel("Correlation coefficient")
plt.tight_layout()
plt.show()
```

```python
#correlation with metacritic score
corrs = corr_matrix['metacritic_score'].sort_values(ascending=False)
corrs.drop('metacritic_score').plot(kind='barh', figsize=(10,6), title="Correlation with Metacritic Score")
plt.xlabel("Correlation coefficient")
plt.tight_layout()
plt.show()
```

```python
#correlation with imdb score
corrs = corr_matrix['imdb_score'].sort_values(ascending=False)
corrs.drop('imdb_score').plot(kind='barh', figsize=(10,6), title="Correlation with IMDB Score")
plt.xlabel("Correlation coefficient")
plt.tight_layout()
plt.show()
```

```python
#Time Series Analysis

#converts the release data to datetime and extracts the year and month
pixar['release_date'] = pd.to_datetime(pixar['release_date'])
pixar['release_year'] = pixar['release_date'].dt.year
pixar['release_month'] = pixar['release_date'].dt.month
```

```python
#graphs the box office worldwide over years
boxOffice_yearly = pixar.groupby('release_year')['box_office_worldwide'].mean()
plt.figure(figsize=(12,6))
sns.lineplot(x=boxOffice_yearly.index, y=boxOffice_yearly.values, marker='o')
plt.title('Average Box Office Worldwide Over Time')
plt.xlabel('Release Year')
plt.ylabel('Avg Worldwide Box Office ($)')
plt.show()
```

```python
#graphs the buget over years
budget_yearly = pixar.groupby('release_year')['budget_clean'].mean()
plt.figure(figsize=(12,6))
sns.lineplot(x=budget_yearly.index, y=budget_yearly.values, marker='o', color='green')
plt.title('Average Budget Over Time')
plt.xlabel('Release Year')
plt.ylabel('Avg Budget ($)')
plt.show()
```

```python
#graphs critic scores over years
avg_scores_by_year = pixar.groupby('release_year').agg({
    'rotten_tomatoes_score': 'mean',
    'metacritic_score': 'mean',
    'imdb_score': 'mean'
}).reset_index()

# Scale RT and Metacritic scores to be out of 10
avg_scores_by_year['rotten_tomatoes_score'] /= 10
avg_scores_by_year['metacritic_score'] /= 10


plt.figure(figsize=(12,6))
sns.lineplot(data=avg_scores_by_year, x='release_year', y='rotten_tomatoes_score', label='Rotten Tomatoes', marker='o')
sns.lineplot(data=avg_scores_by_year, x='release_year', y='metacritic_score', label='Metacritic', marker='s')
sns.lineplot(data=avg_scores_by_year, x='release_year', y='imdb_score', label='IMDB', marker='^')
plt.title('Average Critic Scores Over Time')
plt.xlabel('Release Year')
plt.ylabel('Avg Score')
plt.legend()
plt.show()
```

```python
#graphs box office worldwide over months
boxOffice_monthly = pixar.groupby('release_month')['box_office_worldwide'].mean()
plt.figure(figsize=(12,6))
sns.lineplot(x=boxOffice_monthly.index, y=boxOffice_monthly.values, marker='o')
plt.title('Average Box Office Worldwide Over Time')
plt.xlabel('Release Month')
plt.ylabel('Avg Worldwide Box Office ($)')
plt.show()
```

```python
#graphs budget over months
budget_monthly = pixar.groupby('release_month')['budget_clean'].mean()
plt.figure(figsize=(12,6))
sns.lineplot(x=budget_monthly.index, y=budget_monthly.values, marker='o', color='green')
plt.title('Average Budget Over Time')
plt.xlabel('Release Month')
plt.ylabel('Avg Budget ($)')
plt.show()
```

```python
#graphs critic scores over months
avg_scores_by_month = pixar.groupby('release_month').agg({
    'rotten_tomatoes_score': 'mean',
    'metacritic_score': 'mean',
    'imdb_score': 'mean'
}).reset_index()

# Scale RT and Metacritic scores to be out of 10
avg_scores_by_month['rotten_tomatoes_score'] /= 10
avg_scores_by_month['metacritic_score'] /= 10

plt.figure(figsize=(12,6))
sns.lineplot(data=avg_scores_by_month, x='release_month', y='rotten_tomatoes_score', label='Rotten Tomatoes', marker='o')
sns.lineplot(data=avg_scores_by_month, x='release_month', y='metacritic_score', label='Metacritic', marker='s')
sns.lineplot(data=avg_scores_by_month, x='release_month', y='imdb_score', label='IMDB', marker='^')
plt.title('Average Critic Scores Over Time')
plt.xlabel('Release Month')
plt.ylabel('Avg Score')
plt.legend()
plt.show()
```

Pixar movies generally have high critical reception with Rotten Tomatoes, Metacritic, and IMDB scores all showing strong correlation. Budget and worldwide box office have a weak positive relationship, but higher budgets don't guarantee critical success. The strongest box office correlations were between international and worldwide totals, meaning domestic performance doesn't always match global success. Critical scores have slightly declined over time, especially after 2011, which budgets and box office earnings have generally increased. May and June are peak months for box office returns, especially May. May is also the highest month for critical scores. Movies released in the summer perform better financially while winter releases underperform. There's no strong correlation between run time and any other feature. 

We had challenges with some data fields having missing or placeholder values. This was addressed by converting them to NaN and using new columns to handle outliers and visualize trends more accurately. The sample size was small so it was difficult to draw conclusions for relationships as some trends could have been exaggerated due to outliers. The critic scores being different scales was also a challenge that was solved by putting them all on the same scale.
