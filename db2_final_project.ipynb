```python
#function to connect to server
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
#function to create a database
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
#function to connect to database
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
#function to execute query
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
#Creates the tables for the game
create_user_table = """
CREATE TABLE Users ( 

    user_id INT AUTO_INCREMENT PRIMARY KEY, 

    username VARCHAR(50) NOT NULL, 

    password VARCHAR(255) NOT NULL, 

    high_score INT DEFAULT 0, 

    avg_score FLOAT DEFAULT 0, 

    last_score INT DEFAULT 0, 

    num_times_played INT DEFAULT 0 

); 
"""
create_quiz_table = """
CREATE TABLE Quiz ( 

    question_nbr INT AUTO_INCREMENT PRIMARY KEY, 

    question TEXT NOT NULL, 

    a VARCHAR(255) NOT NULL, 

    b VARCHAR(255) NOT NULL, 

    c VARCHAR(255) NOT NULL, 

    d VARCHAR(255) NOT NULL, 

    correct_answer CHAR(1) NOT NULL 

); 

"""

#connects to the game database
connection = create_db_connection("localhost", "root", pw, "game")

#creates the user and quiz tables
execute_query(connection, create_user_table)
execute_query(connection, create_quiz_table)
```

```python
#function to read results
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

```python
import mysql.connector
from tkinter import Tk, Button, Label, messagebox, simpledialog, StringVar

cursor = connection.cursor(dictionary=True)

# Utility Functions
def fetch_questions():
    #Fetch all questions from the Quiz table in random order
    cursor.execute("SELECT * FROM Quiz ORDER BY RAND()")
    return cursor.fetchall()

def update_user_stats(user_id, score):
    #Update user statistics after a game session
    cursor.execute("SELECT high_score, avg_score, num_times_played FROM Users WHERE user_id = %s", (user_id,))
    user = cursor.fetchone()

    if user:
        high_score = max(user['high_score'], score)
        num_times_played = user['num_times_played'] + 1
        avg_score = ((user['avg_score'] * user['num_times_played']) + score) / num_times_played
        
        cursor.execute("""
            UPDATE Users
            SET high_score = %s, avg_score = %s, last_score = %s, num_times_played = %s
            WHERE user_id = %s
        """, (high_score, avg_score, score, num_times_played, user_id))
        connection.commit()

# Tkinter Application
class GameApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Social Engineering Quiz Game")
        
        self.current_user_id = None
        self.questions = []
        self.current_question_index = 0
        self.score = 0
        
        self.question_var = StringVar()
        self.answer_buttons = []
        
        self.setup_main_menu()

    def setup_main_menu(self):
        #Setup the main menu with buttons for login, account creation, and guest play
        self.clear_window()
        Label(self.root, text="Welcome to the Quiz Game", font=("Arial", 16)).pack(pady=10)
        
        Button(self.root, text="Login", command=self.login).pack(pady=5)
        Button(self.root, text="Create Account", command=self.create_account).pack(pady=5)
        Button(self.root, text="Play as Guest", command=self.play_as_guest).pack(pady=5)
        Button(self.root, text="Exit", command=self.root.quit).pack(pady=5)

    def clear_window(self):
        #Clear all widgets from the window
        for widget in self.root.winfo_children():
            widget.destroy()

    def login(self):
        #Handle user login
        #username = simpledialog.askstring("Login", "Enter your username:")
        #if '#' in username:
                #messagebox.showerror("Error", "Username cannot contain '#'")
                #break
                #self.setup_main_menu()
        #password = simpledialog.askstring("Login", "Enter your password:", show="*")
        #if '#' in password:
                #messagebox.showerror("Error", "Password cannot contain '#'")
                #self.setup_main_menu()
        #cursor.execute("SELECT * FROM Users WHERE username = %s AND password = %s", (username, password))
        #user = cursor.fetchone()
        #if user:
            #self.current_user_id = user['user_id']
            #messagebox.showinfo("Login Success", f"Welcome back, {username}!")
            #self.show_game_or_stats()
        #else:
            #messagebox.showerror("Error", "Invalid username or password.")
            #self.setup_main_menu()

        username = simpledialog.askstring("Login", "Enter your username:")
        if '#' in username:
            messagebox.showerror("Error", "Username cannot contain '#'")
            self.setup_main_menu()
        else:
            # Check if username is valid
            cursor.execute("SELECT user_id FROM Users WHERE username = %s", (username,))
            user = cursor.fetchone()
            if user:
                password = simpledialog.askstring("Login", "Enter your password:", show="*")
                if '#' in password:
                    messagebox.showerror("Error", "Password cannot contain '#'")
                    self.setup_main_mun()
                else:
                    cursor.execute("SELECT * FROM Users WHERE username = %s AND password = %s", (username, password))
                    user = cursor.fetchone()
                    if user:
                        elf.current_user_id = user['user_id']
                        messagebox.showinfo("Login Success", f"Welcome back, {username}!")
                        self.show_game_or_stats()
                    else:
                        messagebox.showerror("Error", "Invalid password.")
                        self.setup_main_menu()

    def create_account(self):
        #Handle new account creation
        while True:
            username = simpledialog.askstring("Create Account", "Enter a username:")
            if '#' in username:
                messagebox.showerror("Error", "Username cannot contain '#'")
                continue
            # Check if username already exists
            cursor.execute("SELECT user_id FROM Users WHERE username = %s", (username,))
            existing_user = cursor.fetchone()
            if existing_user:
                messagebox.showerror("Error", "Username already taken. Please choose another.")

            password = simpledialog.askstring("Create Account", "Enter a strong password (10+ characters):", show="*")
            if len(password) < 10 or '#' in password:
                messagebox.showerror("Error", "Password must be at least 10 characters and cannot contain '#'")
                continue

            try:
                cursor.execute("INSERT INTO Users (username, password) VALUES (%s, %s)", (username, password))
                connection.commit()
                cursor.execute("SELECT user_id FROM Users WHERE username = %s", (username,))
                user = cursor.fetchone()
                self.current_user_id = user['user_id']
                messagebox.showinfo("Success", "Account created successfully!")
                self.show_game_or_stats()
                break
            except mysql.connector.Error as err:
                messagebox.showerror("Error", f"Could not create account: {err}")
                break

    def play_as_guest(self):
        #Start the game in guest mode
        print("Guest play clicked!")  # Debugging
        self.current_user_id = None
        self.show_game_or_stats()

    def show_game_or_stats(self):
        #Give option to play the game or view stats after login or guest play
        print(f"Current User ID: {self.current_user_id}")  # Debugging
        self.clear_window()
        if self.current_user_id:
            # If a user is logged in, show both options
            Button(self.root, text="Play Game", command=self.start_game).pack(pady=5)
            Button(self.root, text="View Stats", command=self.view_stats).pack(pady=5)
        else:
            # If no user is logged in (guest), only show Play Game option
            Button(self.root, text="Play Game as Guest", command=self.start_game).pack(pady=5)

    def start_game(self):
        #Initialize the game and display the first question
        print("Starting game...")  # Debugging
        self.questions = fetch_questions()
        if not self.questions:
            messagebox.showerror("Error", "No questions found in the database!")
            return

        self.current_question_index = 0
        self.score = 0
        self.display_question()

    def display_question(self):
        #Display the current question with answer options
        self.clear_window()
        if self.current_question_index >= len(self.questions):
            self.end_game()
            return

        question = self.questions[self.current_question_index]
        self.question_var.set(question['question'])
        
        Label(self.root, textvariable=self.question_var, font=("Arial", 14), wraplength=400).pack(pady=10)

        # Display answer buttons
        options = ['a', 'b', 'c', 'd']
        for option in options:
            button = Button(self.root, text=f"{option.upper()}. {question[option]}", 
                            command=lambda opt=option: self.check_answer(opt))
            button.pack(pady=5)
            self.answer_buttons.append(button)

    def check_answer(self, selected_option):
        #Check if the selected answer is correct
        question = self.questions[self.current_question_index]
        if selected_option == question['correct_answer']:
            messagebox.showinfo("Correct", "Your answer is correct!")
            self.score += 1
        else:
            messagebox.showinfo("Incorrect", f"The correct answer was {question['correct_answer']}.")

        self.current_question_index += 1
        self.display_question()

    def end_game(self):
        #End the game and update the database if necessary
        total_questions = len(self.questions)
        messagebox.showinfo("Game Over", f"Your final score is {self.score} out of {total_questions}.")
        if self.current_user_id:
            update_user_stats(self.current_user_id, self.score)

        self.setup_main_menu()

    def view_stats(self):
        #View user stats
        cursor.execute("SELECT * FROM Users WHERE user_id = %s", (self.current_user_id,))
        user = cursor.fetchone()

        stats_window = Tk()
        stats_window.title("User Stats")

        stats_label = Label(stats_window, text=f"Username: {user['username']}\nHigh Score: {user['high_score']}\nAverage Score: {user['avg_score']:.2f}\nLast Score: {user['last_score']}\nTimes Played: {user['num_times_played']}")
        stats_label.pack()

# Run the application
if __name__ == "__main__":
    root = Tk()
    app = GameApp(root)
    root.mainloop()
```
